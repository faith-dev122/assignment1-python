```python
#Factorial Program
def factorial(n):
    result = 1
    for i in range(1, n+1):
        result *= i
    return result

number = int(input("Enter a number: "))
print("Factorial:", factorial(number))
```

```python
#Fibonacci Program
a, b = 0, 1
for _ in range(10):
    print(a, end=' ')
    a, b = b, a + b
```

```python
#Quadratic Roots Program
import math
a = float(input("Enter a: "))
b = float(input("Enter b: "))
c = float(input("Enter c: "))
discriminant = b**2 - 4*a*c

if discriminant >= 0:
    root1 = (-b + math.sqrt(discriminant)) / (2*a)
    root2 = (-b - math.sqrt(discriminant)) / (2*a)
    print("Roots:", root1, root2)
else:
    print("Complex roots")
```

```python
#Volume under z = x² + y²
step = 0.1
volume = 0.0

x = 0.0
while x <= 1.0:
    y = 0.0
    while y <= 1.0:
        z = x**2 + y**2
        volume += z * step * step
        y += step
    x += step

print("Approximate volume:", volume)
```
